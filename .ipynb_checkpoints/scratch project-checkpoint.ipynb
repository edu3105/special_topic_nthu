# Gerchberg Saxton


```python
def gerchberg_saxton(input_intensity, target_intensity, num_iterations):
    # Initialize phase distribution randomly
    phase = np.exp(1j * 2 * np.pi * np.random.rand(*input_intensity.shape))
    
    #print(phase)

    for _ in range(num_iterations):
        # Fourier transform to the output plane
        output_field = fft2(np.sqrt(input_intensity) * np.exp(1j * phase))
        
        # Replace amplitude with target intensity
        output_amplitude = np.sqrt(target_intensity) * np.exp(1j * np.angle(output_field))
        
        # Inverse Fourier transform to the input plane
        input_field = ifft2(output_amplitude)
        
        # Update phase
        phase = np.angle(input_field)

    return np.abs(input_field)**2
```

# Iterative Fourier Transform Algorithm (IFTA)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft2, ifft2

def gerchberg_saxton(input_intensity, target_intensity, num_iterations):
    # Ensure input and target are arrays with the same shape
    input_intensity = np.asarray(input_intensity, dtype=np.complex128)
    target_intensity = np.asarray(target_intensity, dtype=np.complex128)

    # Initialize phase distribution randomly
    phase = np.exp(1j * 2 * np.pi * np.random.rand(*input_intensity.shape))

    for _ in range(int(num_iterations)):  # Ensure num_iterations is an integer
        # Fourier transform to the output plane
        output_field = fft2(np.sqrt(input_intensity) * np.exp(1j * phase))
        
        # Replace amplitude with target intensity
        output_amplitude = np.sqrt(target_intensity) * np.exp(1j * np.angle(output_field))
        
        # Inverse Fourier transform to the input plane
        input_field = ifft2(output_amplitude)
        
        # Update phase
        phase = np.angle(input_field)

    return np.abs(input_field)**2

def cost_function(iterations, input_intensity, target_intensity):
    result = gerchberg_saxton(input_intensity, target_intensity, num_iterations=iterations)
    return np.sum((result - target_intensity)**2)

# Example usage:
size = 256
input_intensity = np.random.rand(size, size)
target_intensity = np.ones((size, size))

# Gradient Descent for Iterations
learning_rate = 0.1
current_iterations = 50.0  # Initial guess for the number of iterations

# Number of iterations to perform gradient descent
num_gradient_descent_iterations = 10

for _ in range(num_gradient_descent_iterations):
    # Calculate gradient using finite differences
    cost_current = cost_function(current_iterations, input_intensity, target_intensity)
    cost_next = cost_function(current_iterations + 1, input_intensity, target_intensity)
    
    gradient = (cost_next - cost_current) / 1  # Finite differences
    
    # Update the number of iterations using gradient descent
    current_iterations -= learning_rate * gradient

# Final result using the optimized number of iterations
result = gerchberg_saxton(input_intensity, target_intensity, num_iterations=current_iterations)

# Plot the results
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(input_intensity, cmap='gray')
plt.title('Input Intensity')

plt.subplot(1, 3, 2)
plt.imshow(target_intensity, cmap='gray')
plt.title('Target Intensity')

plt.subplot(1, 3, 3)
plt.imshow(result, cmap='gray')
plt.title(f'Output Intensity (Optimal Iterations: {int(current_iterations)})')

plt.show()
```

# Flat Lorentzian

this is just trying to apply the function given in the paper

```python
def flatLorentzian(r, Rfl, q):
    #last equation
    
    pi = np.pi
    return 1/ (pi * Rfl**2 * (1 + (r / Rfl)**q)**(1 + 2 / q))
```

```python
def topHatBeam(size, Rfl, q):
    #2d array for top hat beam
    
    x, y = np.meshgrid(np.linspace(-1, 1, size), np.linspace(-1, 1, size))
    radialCoordinates = np.sqrt(x**2 + y**2)
    intensityProfile = flatLorentzian(radialCoordinates, Rfl, q)
    return intensityProfile.astype(float)
```

```python
def plotIntensity(intensityProfile):
    # plotting 2d intensity profile
    
    plt.imshow(intensityProfile, cmap='viridis', extent=[-1, 1, -1, 1])
    plt.colorbar(label='Intensity')
    plt.title('Modified Flattened-Lorentzian Top-Hat Beam')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()
```

```python
#for usage
size = 200
Rfl = 0.5
q = 2.0
```

```python
intensityProfile = topHatBeam(size, Rfl, q)
plotIntensity(intensityProfile)
```
